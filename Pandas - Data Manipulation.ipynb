```python
pd.Categorical(titanic_data['Pclass'])  # it shows, what are the categorical values available in Pclass(passenger class)
```

```python
# Now renaming the category present in Pclass
new_Pclass = pd.Categorical(titanic_data['Pclass'])
new_Pclass = new_Pclass.rename_categories(['c1','c2','c3'])
new_Pclass
```

```python
titanic_data["Pclass"] = new_Pclass  # the new category for Pclass is changed in the titanic_data set
```

```python
# renaming the categorgies in Sex column 
c_sex = pd.Categorical(titanic_data['Sex'])
c_sex=c_sex.rename_categories(['Male','Female'])
c_sex
```
